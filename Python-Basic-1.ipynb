### Variables and Data Types

#### Variables hold data for the program

#### Python automatically assigns data types when you assign values (dynamic typing).

##### Press (Shift + Enter) to execute the cell

```python
var_int = 25
var_flt = 25.49
var_str = "Python"
var_bol = True

#Check the type of variable
type(var_str)

print("Hello World","This is Python class", 500, True)
print("Type information for this variable is", type(var_bol))

#f-strings (formatted string literals) are modern and efficient way to include 
# variables or expressions directly inside a string

print(f"Type information for this variable is {var_str}")

print(var_int, var_flt, var_str, var_bol)

print("Type information of each of the variables")
print("Type of variable var_int is", type(var_int))
print("Type of variable var_flt is", type(var_flt))
print("Type of variable var_str is", type(var_str))
print("Type of variable var_bol is", type(var_bol))
```

#### More Data Types

#### Lists are collection of mutable items which means after variable creation you can (add, remove, update) elements

```python
list_colors = ["red", "green", "blue","green"]
print("Type of the list_colors is as follows: ", type(list_colors))
print(list_colors)
```

#### Tuples are collection of immutable items â€” you cannot modify, add, or remove elements after creation.

```python
tuple_datapoints = (10, 20, 20) 
print("Type of the tuple_datapoints is as follows: ", type(tuple_datapoints))
print(tuple_datapoints)
```

#### A dictionary in Python is a collection of key-value pairs. Each key acts as a unique identifier, and each value stores the data related to that key

```python
sample_dict = {
    "key1": "Java",
    "key2": "Python",
    "key3": "React",
    "key4": "Angular"
}
print(sample_dict)
print(type(sample_dict))
```

#### A set is an unordered collection of unique elements.

```python
fruits = {"apple", "banana", "mango", "apple"}
print(fruits)
```

```python
set2 = set(["apple", "banana", "cherry"])
print(set2)
```

#### CRUD operations for the List

```python
# Create a list
#        [ 0,         1,          2]
fruits = ["apple", "banana", "cherry"]
print("Original list:", fruits)

# Modify element
fruits[1] = "mango"
print("After modification:", fruits)

# Add element
fruits.append("orange")
print("After adding element:", fruits)

# Remove element
fruits.remove("apple")
print("After removing element:", fruits)
```

#### CRUD operations for the Dictionary 

```python
# Add new key-value pairs
#<dictinary_name>[<key>] = <value>

sample_dict["key5"] = "NodeJS"
sample_dict["key6"] = "Django"

print("After New Entries:", sample_dict)
```

```python
# Access individual value
# Dictionary-variable["your-key"]
print(sample_dict["key2"])

# Using get() to avoid errors if key doesn't exist
print(sample_dict.get("key3"))
print(sample_dict.get("key10", "Not Found")) 

# Loop through dictionary
print("\nAll keys and values:")
for key, value in sample_dict.items():
    print(key, ":", value)
```

```python
# Update a single key
sample_dict["key1"] = "JavaScript"

# Update multiple keys
#sample_dict.update({"key2": "Python 3", "key4": "Angular 17"})

sample_dict.update({"key2": "Python 3", "key4": "Angular 17", "key7": "Angular 77"})

print("After UPDATE:", sample_dict)
```

```python
# Delete specific key
del sample_dict["key5"]
print("After deleting key5:", sample_dict)

# Using pop() removes and returns the value
removed_value = sample_dict.pop("key6")
print("Removed:", removed_value)
print("After POP:", sample_dict)

# Remove last inserted item (Python 3.7+)
sample_dict.popitem()
print("After popitem():", sample_dict)

# Clear all items
sample_dict.clear()
print("After CLEAR:", sample_dict)
```
