```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
X = np.array([[1],[2],[3],[4]])   # Hours studied
y = np.array([40, 50, 60, 70])    # Marks scored

# Model
model = LinearRegression()
model.fit(X, y)

# Output
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

```python
# Predict for 5 hours
print("Prediction for 5 hours:", model.predict([[5]])[0])
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

# Features: [hours_studied, hours_slept]
X = [
    [1, 6],
    [2, 5],
    [3, 6],
    [4, 6],
    [5, 7],
    [6, 5],
]

# Labels: 0 = Fail, 1 = Pass
y = [0, 0, 0, 1, 1, 1]

model = RandomForestClassifier(n_estimators=10, random_state=42)
model.fit(X, y)

print(model.predict([[4, 6]]))   # Predict for new student
```

```python
from sklearn.ensemble import RandomForestRegressor

# Feature: house size (sq ft)
X = [
    [500],
    [800],
    [1000],
    [1200],
    [1500]
]

# Target: price
y = [5, 8, 10, 12, 15]   # in lakhs (example)

model = RandomForestRegressor(n_estimators=20, random_state=42)
model.fit(X, y)

print(model.predict([[1100]]))
```

##### What is XGBoost?

XGBoost = eXtreme Gradient Boosting

It is a powerful tree-based machine learning algorithm that builds models step-by-step and improves errors at each step.

Think of it like this:

Start with a simple model → see mistakes → build another model to fix those mistakes → repeat.

In the end, all small models combine into one very strong model.

XGBoost is an improved, super-optimized version of Gradient Boosting.

Why is it called “boosting”?

Because each new model boosts (improves) performance by correcting previous mistakes.

Random Forest = many trees built independently
XGBoost = trees built sequentially, learning from mistakes
