#### Ploting charts using Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
avg_salary = df.groupby("department")["salary"].mean()

plt.figure(figsize=(8,5))
plt.bar(avg_salary.index, avg_salary.values)
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.xticks(rotation=45)
plt.show()
```

```python
df["join_date"] = pd.to_datetime(df["join_date"])
df_sorted = df.sort_values("join_date")

plt.figure(figsize=(10,5))
plt.plot(df_sorted["join_date"], df_sorted["salary"])
plt.title("Salary Over Join Dates")
plt.xlabel("Join Date")
plt.ylabel("Salary")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df["age"], df["salary"])
plt.title("Salary vs Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df["salary"], bins=5)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
dept_counts = df["department"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(dept_counts, labels=dept_counts.index, autopct="%1.1f%%")
plt.title("Department Distribution")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))

# Plot 1
ax[0].bar(df["city"].value_counts().index,
          df["city"].value_counts().values)
ax[0].set_title("Employees per City")

# Plot 2
ax[1].hist(df["age"], bins=5)
ax[1].set_title("Age Distribution")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.boxplot(column="salary", by="department")
plt.title("Salary Box Plot by Department")
plt.suptitle("")  # removes default extra title
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.barh(df["name"], df["experience_years"])
plt.title("Experience Years of Employees")
plt.xlabel("Experience (years)")
plt.ylabel("Employee Name")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df["age"], df["salary"])

for i in range(len(df)):
    if df["salary"][i] > 90000:
        plt.text(df["age"][i], df["salary"][i], df["name"][i])

plt.title("High Earning Employees Highlighted")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

#### Pydantic 

Pydantic is a Python library for the following:

Data validation
Data parsing
Automatic type checking
Creating clean data models (like DTOs)

It is heavily used in:

-- FastAPI
-- Backend validation
-- ML model configs
-- JSON parsing
-- Microservices

Pydantic ensures:
"Your data is correct before your code runs."

```python
from pydantic import BaseModel

class User(BaseModel):
    name: str
    age: int
    email: str
```

```python
u = User(name="Keerthi", age=33, email="keerthi@gmail.com")
print(u)
```
